## Subset & Data Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_subset = df[['pH', 'quality']]
```

```python
ss = StandardScaler()
```

```python
x_scaled = ss.fit_transform(df_subset)
```

# KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
model_kmeans = KMeans(n_clusters = 3)
```

```python
model_kmeans.fit(x_scaled)
```

```python
print(x_scaled)
```

```python
labels = model_kmeans.labels_
```

```python
print(labels)
```

## membuat cross_tab untuk membandingkan antar cluster

```python
# buat dataframe dengan label dan species dari iris sebagai kolom
df_kmeans = pd.DataFrame({'labels' : labels, 'type': df.type})
```

```python
# Buat Crosstab
ct1 = pd.crosstab(df_kmeans['labels'], df_kmeans['type'])
```

```python
print(ct1)
```

## Melihat Centroid

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
# untuk mengambil semua baris pad kolom tertentu (berbentuk array)
xs = x_scaled[:, 0]
ys = x_scaled[:,1]
```

```python
#menampung koordinat dari semua centroid
centroids = model_kmeans.cluster_centers_
```

```python
# pisahkan centroid kordinat x dan kordinat y
centroid_x = x_scaled[:, 0]
centroid_y = x_scaled[:,1]
```

```python
# label aktual
df_aktual = df[["fixed acidity", "citric acid", "type"]]
sns.scatterplot(x = "fixed acidity", y = "citric acid", data = df_aktual, hue = "type")
plt.show()
```

```python
# Visualisasi Cluster
plt.figure(figsize = (5,5))

# Titik-titik data x dan y tiap cluster
plt.scatter(xs, ys, alpha = 0.5, c = labels)

# Titik-titik centroid
plt.scatter(centroid_x, centroid_y, label = "Centroid", s=50,  marker = "^")

# Beri Label 
plt.xlabel("foxed acidity")
plt.ylabel("citric acid")

plt.legend()
plt.show()
```

```python
ks = range(1, 6)
inertias = []
for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(x_scaled)
    inertias.append(model.inertia_)
# Plot ks vs inertias
plt.figure(figsize=(5,5))
plt.plot(ks, inertias, '-o')
plt.xlabel('Jumlah Cluster, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
# Memvisualkan hasil kluster
plt.figure
sct = plt.scatter(x_scaled[:,0], x_scaled[:,1], s = 50,
                  c = labels, marker = "o", alpha = 0.5)
plt.scatter(centroid_x, centroid_y, c='blue', s=200, alpha=0.5);
plt.title("Hasil Klustering K-Means")
plt.xlabel("fixed acidity")
plt.ylabel("citric acid")
plt.show()
```
