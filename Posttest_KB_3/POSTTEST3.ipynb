```python
data_pie = dataFrame['characters'].value_counts()
data_pie.head()
```

```python
data_pie = dataFrame['characters'].value_counts().rename_axis('characters')
data_pie.head()
```

```python
data_pie = dataFrame['episodeNum'].value_counts().rename_axis('episodeNum').reset_index(name = 'id_count')
```

```python
plt.pie(data_pie.id_count,
        labels = data_pie['episodeNum'],
        startangle=180,
        autopct='%.1f%%',
        shadow = True
       )
plt.title('jumlah episode pada dataset')
plt.show()        
```
